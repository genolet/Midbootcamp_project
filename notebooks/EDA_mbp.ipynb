```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats
import math
import scipy.stats as st
from scipy.stats import ttest_ind 
```

```python
listings = pd.read_csv('/Users/origenolet/Desktop/Ironhack/Week5/Mid-bootcamp-project/data/clean/listings_clean.csv')
display(listings.head())
```

```python
### I will create a dic to save all the generated pvalues 

dic_pval = {}
```

```python
print(listings["n_bathrooms"].value_counts())

baths_0 = listings[listings["n_bathrooms"] == 0]
baths_0 = baths_0["price"]
baths_0 = list(baths_0)

baths_4 = listings[listings["n_bathrooms"] == 4]
baths_4 = baths_4["price"]
baths_4 = list(baths_4)

print(np.var(baths_0), np.var(baths_4))
x = ttest_ind(baths_0, baths_4, equal_var=False)

dic_pval["n_bathrooms"] = x[1]
print(dic_pval)
```

```python
print(listings["accommodates"].value_counts())

acco_1 = listings[listings["accommodates"] == 1]
acco_1 = acco_1["price"]
acco_1 = list(acco_1)

acco_16 = listings[listings["accommodates"] == 16]
acco_16 = acco_16["price"]
acco_16 = list(acco_16)

print(np.var(acco_1), np.var(acco_16))
x = ttest_ind(acco_1, acco_16, equal_var=False)

dic_pval["accomodates"] = x[1]
print(dic_pval)
```

```python
print(listings["bedrooms"].value_counts())

bedr_1 = listings[listings["bedrooms"] == 1]
bedr_1 = bedr_1["price"]
bedr_1 = list(bedr_1)

bedr_7 = listings[listings["bedrooms"] == 7]
bedr_7 = bedr_7["price"]
bedr_7 = list(bedr_7)

print(np.var(bedr_1), np.var(bedr_7))
x = ttest_ind(bedr_1, bedr_7, equal_var=False)

dic_pval["bedrooms"] = x[1]
print(dic_pval)
```

```python
display(listings.groupby(["district"]).agg({"price":"median"}))

district_mitte = listings[listings["district"] == "Mitte"]
district_mitte = district_mitte["price"]
district_mitte = list(district_mitte)

district_neukölln = listings[listings["district"] == "Neukölln"]
district_neukölln = district_neukölln["price"]
district_neukölln = list(district_neukölln)

print(np.var(district_mitte), np.var(district_neukölln))
x = ttest_ind(district_mitte, district_neukölln, equal_var=False)

dic_pval["district"] = x[1]
print(dic_pval)
```

```python
df = listings.groupby(["neighbourhood"]).agg({"price":"median"})
display(df.sort_values(by = ["price"], ascending=False))

neighbourhood_ah = listings[listings["neighbourhood"] == "Alt-Hohenschönhausen Süd"]
neighbourhood_ah = neighbourhood_ah["price"]
neighbourhood_ah = list(neighbourhood_ah)

neighbourhood_bm = listings[listings["neighbourhood"] == "Baumschulenweg"]
neighbourhood_bm = neighbourhood_bm["price"]
neighbourhood_bm = list(neighbourhood_bm)

print(np.var(neighbourhood_ah), np.var(neighbourhood_bm))
x = ttest_ind(neighbourhood_ah, neighbourhood_bm, equal_var=True)

dic_pval["neighbourhood"] = x[1]
print(dic_pval)
```

```python
df = listings.groupby(["room_type"]).agg({"price":"median"})
display(df.sort_values(by = ["price"], ascending=False))

room_hotel = listings[listings["room_type"] == "Hotel room"]
room_hotel = room_hotel["price"]
room_hotel = list(room_hotel)

room_shared = listings[listings["room_type"] == "Shared room"]
room_shared = room_shared["price"]
room_shared = list(room_shared)


print(np.var(room_hotel), np.var(room_shared))
x = ttest_ind(room_hotel, room_shared, equal_var=True)

dic_pval["room_type"] = x[1]
print(dic_pval)
```

```python
df = listings.groupby(["property_type"]).agg({"price":"median"})
display(df.sort_values(by = ["price"], ascending=False))

room_apart = listings[listings["property_type"] == "Room in serviced apartment"]
room_apart = room_apart["price"]
room_apart = list(room_apart)

room_shared = listings[listings["property_type"] == "Shared room in rental unit"]
room_shared = room_shared["price"]
room_shared = list(room_shared)


print(np.var(room_apart), np.var(room_shared))
x = ttest_ind(room_apart, room_shared, equal_var=False)

dic_pval["property_type"] = x[1]
print(dic_pval)
```

```python
cols = ["balcony", "bbq", "parking", "washer", "wifi", "kitchen", "host_email", "host_work_email"]

for col in cols:
    feature1 = listings[listings[col] == 1]
    feature1 = feature1["price"]
    feature1 = list(feature1)
    feature2 = listings[listings[col] == 0]
    feature2 = feature2["price"]
    feature2 = list(feature2)
    print("T-test for:", col)
    print(np.var(feature1), np.var(feature2))
    x = ttest_ind(feature1, feature2, equal_var=True)
    dic_pval[col] = x[1]
    print(dic_pval)
```
